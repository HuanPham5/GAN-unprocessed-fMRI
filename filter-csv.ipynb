```python
import pandas as pd
import random

# Use demographic csv
data = pd.read_csv('../GAN-neural-imaging/csvfiles/1139_Demographics_20150429.csv')

# Display the first few rows of the DataFrame
data.head()
```

```python
data.dtypes
```

```python
data.columns
```

```python
# How many patients of each type?
unique_values_count = data['Unnamed: 1'].value_counts()
unique_values_count
```

```python
schizophrenia_count = (data['Unnamed: 1'] == 'Schizophrenia').sum()
schizophrenia_count
```

```python
# Filter rows where 'Unnamed: 1' column contains 'Schizophrenia' and extract the 'Unnamed: 0' column
schizophrenia_ids = data.loc[data['Unnamed: 1'] == 'Schizophrenia', 'Unnamed: 0']

# Convert the result to a list
schizophrenia_id_list = schizophrenia_ids.tolist()

# Print the list of IDs
print("IDs of individuals with schizophrenia:")
for id in schizophrenia_id_list:
    print(f" '{id}' ,")

print("how many schizophrenic?" ,len(schizophrenia_id_list))
```

```python
print("IDs of individuals with schizophrenia: ", end="")
print(", ".join(f"'{id}'" for id in schizophrenia_id_list))
```

```python
# Take a random subset of n IDs
random_subset = random.sample(schizophrenia_id_list, 10)

# Print the random subset of IDs
print("Random subset of 10 IDs of individuals with schizophrenia:")
for id in random_subset:
    print(id)
```

```python
non_schizophrenia_count = (data['Unnamed: 1'] != 'Schizophrenia').sum()
non_schizophrenia_count
```

```python
# non-schizphrenic include some other types as well

non_schizophrenia_ids = data.loc[data['Unnamed: 1'] != 'Schizophrenia', 'Unnamed: 0']

non_schizophrenia_id_list = non_schizophrenia_ids.tolist()

# Print the list of IDs
print("IDs of individuals with non_schizophrenia:")
for id2 in non_schizophrenia_id_list:
     print(f" '{id2}' ,")
```

```python
print("IDs of individuals with non_schizophrenia: ", end="")
print(", ".join(f"'{id2}'" for id2 in non_schizophrenia_id_list))
```

```python
control_ids = data.loc[data['Unnamed: 1'] == 'Control', 'Unnamed: 0']

control_id_list = control_ids.tolist()

# Print the list of IDs
print("IDs of individuals with control:")
for id3 in control_id_list:
    print(id3)

print("how many controls?" ,len(control_id_list))
```

```python
print("IDs of individuals with control: ", end="")
print(", ".join(f"'{id3}'" for id3 in control_id_list))
```
